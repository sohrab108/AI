```python
import tensorflow as tf
import keras
import random
import gym
print(tf.__version__)   
print(keras.__version__)   
```

```python
env=gym.make('CartPole-v1')
```

```python
actions=env.action_space.n
states=env.observation_space.shape[0]
print(actions)
print(states)
```

```python
run=True
env.reset()
ic=0
while run:
    
    env.render()
    act=random.randint(0,1)
    n_steate,reward,done,info=env.step(act)
    print(n_steate,reward,done,info)
    ic+=1
    if ic>600:
        ic=0
        run=False
env.close()
    
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense,Flatten
from keras.optimizers import Adam
```

```python
modelq=Sequential()
modelq.add(Flatten(input_shape=(1,4)))
modelq.add(Dense(48,activation='relu'))
modelq.add(Dense(24,activation='relu'))
modelq.add(Dense(2,activation='linear'))
modelq.summary()
```

```python
from rl.agents import DQNAgent
from rl.memory import SequentialMemory
from rl.policy import BoltzmannQPolicy
```

```python
policy=BoltzmannQPolicy()
memory=SequentialMemory(limit=50000,window_length=1)
robot=DQNAgent(model=modelq, policy=policy,memory=memory,nb_actions=2
               ,nb_steps_warmup=10,target_model_update=1e-2)
robot.compile(Adam(lr=1e-3),metrics=['mae'])
```

```python
robot.fit(env=env,nb_steps=50000,visualize=False,verbose=True)
```

```python
out=robot.test(env,nb_episodes=10,visualize=False)
```

```python
robot.save_weights('rl.h5')
```
